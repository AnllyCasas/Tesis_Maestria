**GENERANDO DF CON VALORES FALTANTES MENOR IGUAL A 40%**

```python
def get_columns_with_nas_less_than_40(df1):
    missing_values_percentage = df1.isnull().mean() * 100
    if missing_values_percentage.any():
        columns_with_nas_less_than_40 = missing_values_percentage[missing_values_percentage <= 40].index.tolist()
        return columns_with_nas_less_than_40
    else:
        return None
```

```python
columns_with_nas_less_than_40 = get_columns_with_nas_less_than_40(df1)
columns_with_nas_less_than_40
```

```python
def create_new_dataframe_with_columns_less_than_40_nas(df1, columns_with_nas_less_than_40):
    """Creates a new DataFrame from the given DataFrame that only includes the columns that have less than 30% NA values."""
    new_df = df1.loc[:, columns_with_nas_less_than_40]
    return new_df
```

```python
# Create a new DataFrame with the columns with less than 30% NA values
new_df = create_new_dataframe_with_columns_less_than_40_nas(df1, columns_with_nas_less_than_40)

# Print the new DataFrame
print(new_df)
```

```python
import numpy as np
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

##imputaciòn

# Fit the IterativeImputer to the NumPy array
imp = IterativeImputer(max_iter=10, random_state=0)
imp.fit(new_df)

# Impute the missing values in the NumPy array
imputed_new_df_array = imp.transform(new_df)

imputed_new_df = pd.DataFrame(imputed_new_df_array, columns=new_df.columns)

# Print the imputed DataFrame
print(imputed_new_df)
```

```python
def formatear_numero_en_columna(columna):
    # Convertir cada elemento de la columna a cadena para manipular los caracteres
    columna_str = columna.astype(str)

    # Verificar si el punto decimal ya está presente en cada elemento
    sin_punto_decimal = ~columna_str.str.contains('\.')

    # Tomar los tres primeros dígitos solo para los elementos sin punto decimal
    tres_primeros_digitos = columna_str[sin_punto_decimal].str[:3]

    # Resto de la cadena después de los tres primeros dígitos solo para los elementos sin punto decimal
    resto = columna_str[sin_punto_decimal].str[3:]

    # Agregar un punto decimal después de los tres primeros dígitos solo para los elementos sin punto decimal
    columna_formateada = tres_primeros_digitos + '.' + resto

    # Mantener los elementos con punto decimal sin cambios
    columna_formateada = pd.concat([columna_str[~sin_punto_decimal], columna_formateada])

    return columna_formateada

```

```python
imputed_new_df['COLES_nueva'] = formatear_numero_en_columna(imputed_new_df['COLES'])

# Imprimir el DataFrame con la nueva columna formateada
print(imputed_new_df)
```
